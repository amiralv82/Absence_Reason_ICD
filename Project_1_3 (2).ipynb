<b> 2. تحلیل مصرف کنندگان الکل و سیگار </b>
<p align="justify">
<font size=5>
با تجزیه و تحلیل تأثیر مصرف الکل یا دخانیات بر غیبت کارکنان شروع کنیم. با توجه به اینکه سیگار کشیدن و مصرف مکرر الکل تأثیر منفی بر شرایط سلامتی دارند، انتظار داریم که برخی از بیماری‌ها در سیگاری‌ها و مصرف‌کنندگان الکل بیشتر از دیگران باشند.
<br>
1.2. بررسی کنید چند درصد از غیبت ها،از گروه مصرف کننده الکل و چند درصد سیگاری هستند.
</p>
</font>
</div>
</font>

```python
# Calculate the number of alcohol consumers
alcohol_consumers = df[df['Social_Drinker_Status'] == "Yes"]

# Calculate the number of smokers
smokers = df[df['Social_Smoker_Status'] == "Yes"]

# Calculate the number of absences for alcohol consumers
absences_alcohol_consumers = alcohol_consumers['Absence_Reason'].count()

# Calculate the number of absences for smokers
absences_smokers = smokers['Absence_Reason'].count()

# Calculate the total number of absences
total_absences = df['Absence_Reason'].count()

# Calculate the percentage of absences for alcohol consumers and smokers
percentage_absences_alcohol_consumers = (absences_alcohol_consumers / total_absences) * 100
percentage_absences_smokers = (absences_smokers / total_absences) * 100

# Display the results
print("Percentage of absences among alcohol consumers:", percentage_absences_alcohol_consumers)
print("Percentage of absences among smokers:", percentage_absences_smokers)
```

2.2. با توجه به درصد قابل توجه مصرف کنندگان الکل، با استفاده از barplot توزیع دلایل غیبت را در میان مصرف کنندگان الکل نشان دهید.
همچنین  با تحلیل بصری این نمودار بررسی کنید که بیشترین اختلاف بین مصرف کنندگان الکل و سایرین در غیبت کاری، مرتبط با چه دلیلی است.
</p>

```python
# Calculate the number of absences for each reason among alcohol consumers
absence_reason_counts_alcohol_consumers = alcohol_consumers['Absence_Reason'].value_counts()
# Calculate the percentage of each absence reason among alcohol consumers
percentage_absence_reasons_alcohol_consumers = (absence_reason_counts_alcohol_consumers / absences_alcohol_consumers) * 100

# Plot the barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=percentage_absence_reasons_alcohol_consumers.index, y=percentage_absence_reasons_alcohol_consumers.values)
plt.title('Distribution of Absence Reasons among Alcohol Consumers')
plt.xlabel('Absence Reason')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<b> <font color='red'> توضیحات پاسخ: </font></b>

استفاده از نمودار ویولین در موارد زیر می‌تواند مفید باشد:
۱. نمایش توزیع داده‌ها: نمودار ویولین به خوبی نشان می‌دهد که چگونه داده‌ها در طول محور مقیاس شده توزیع شده‌اند.

۳. نمایش اطلاعات مرتبط: با استفاده از نمودار ویولین می‌توان اطلاعات اضافی مانند میانگین، واریانس و کوارتیل‌ها را نیز نمایش داد.

```python
from scipy.stats import yeojohnson
# Apply Yeo-Johnson transformation to absenteeism hours column
df['Absent_Hours_Transformed'], _ = yeojohnson(df['Absence_Duration_Hours'] + 1)  # Adding 1 to handle zero values
```

```python
# Create a joint plot for the relationship between distance to work and absenteeism hours (transformed)
plt.figure(figsize=(10, 6))
sns.jointplot(data=df, x='Distance_to_Work', y='Absent_Hours_Transformed', kind='scatter')
plt.xlabel('Distance to Work')
plt.ylabel('Absent Hours (Transformed)')
plt.show()
```

با توجه به نمودار رسم شده که در آن ساعت های غیبت در ماه های سال بررسی شده است بیشترین میزان غیبت به ترتیب از زیاد به کم در ماه های چهارم سپس سوم سپس ششم سپس هفتم سپس دوازدهم سپس یازدهم سپس پنجم سپس نهم سپس هشتم سپس دهم سپس یکم سپس دوم مشاهده میشود که نشان دهنده ی این موضوع است که در ماه های ابتدایی سال در ماه اول نسبت به ماه دوم غیبت ها بیشتر است ولی از ماه دوم به بعد از ماه سوم تا چهارم شاهد افزایش غیبت ها هستیم که در ماه چهارم غیبت ها به اوج خود می رسد در ماه پنجم غیبت ها کاهش داشته ولی در ماه ششم افزایش می یابد در ماه های هفتم و هشتم غیبت ها سیر نرولی داشته و در ماه  نهم به میزان جزییی غیبت ها افزایش داشته ولی مجددا در ماه دهم غیبت ها به میزان کمی کاهش داشته و پس از ماه دهم در ماه های یازدهم و دوازدهم شاید سیر صعودی و افزایش غیبت ها هستیم.
که این افزایش در ماه های اخر سال می تواند مرتبط با فصل هم باشد زیرا که فصل سرد سال است و میتواند مرخصی ناشی از بیماری را افزایش دهد.
